```python
def ignore_velocity_at_post(X, y):
    new_X = []

    set_i = set()
    for k, x in tqdm(enumerate(X)):
        for i, element in enumerate(x[::-1]):
            if i == len(x) - 1:
                new_X.append(x)
                set_i.add(k)
            elif element[-1] != 0:
                new_X.append(x[:len(x) - i])
                set_i.add(k)
                break

    total_set = set(range(len(X)))

    diff = list(total_set.difference(set_i))[::-1]
    diff.sort()
#     for i in diff[::-1]:
#         del y[i]
    print(diff)
    
    return new_X, y
```

```python

def read_data(folder):
    i = 0

    features_path = os.path.join(folder, 'features')
    for file in tqdm(os.listdir(features_path)):
        if "csv" not in file:
            continue
        new_df = pd.read_csv(os.path.join(features_path, file))
        if i == 0:
            features = new_df
        else:
            features = pd.concat([features, new_df])
        i += 1

    labels = pd.read_csv(os.path.join(folder, 'labels/part-00000-e9445087-aa0a-433b-a7f6-7f4c19d78ad6-c000.csv'))

    return features, labels
```
